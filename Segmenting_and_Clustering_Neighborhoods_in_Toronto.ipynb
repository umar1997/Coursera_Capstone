### Importing Libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

### Data Scraping

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
source = requests.get(url).text
Canada_data = BeautifulSoup(source, 'lxml')
Canada_data = Canada_data.find('table')
Canada_data
```

### Data Cleaning, Wrangling and Transformation

```python
entries = Canada_data.findAll('td')
table = []
for i, entry in enumerate(entries): 
    line = re.split(r'<[^>]*>',str(entry))
    row = []
    for w in line:
        w = w.replace('  ','').replace('(','').replace(')','')
        if '/' in w:
            extras = w.split('/')
            for e in extras:
                if len(e) > 2:
                    row.append(e.strip())
        elif len(w) > 2 and w!= ' / ':
            row.append(w.strip())
    table.append(row)
table
```

```python
First = []
Second = []
Third = []
for row in table:
    if len(row) > 2:
        First.append(row[0])
        Second.append(row[1])
        s = ""
        for w in row[2:]:
            s+= w + ", "
        Third.append(s)
        
            
            
        
df = pd.DataFrame(list(zip(First, Second, Third)), 
               columns =['PostalCode', 'Borough', 'Neigborhood'])
df
```
